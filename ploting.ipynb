```python
import pandas as pd
import plotly.express as px
import sqlite3
```

```python
conn=sqlite3.connect('china_mof.db')
df=pd.read_sql_query('SELECT * FROM financial_data', conn)
conn.close()
df
```

```python
chart_data=df.copy()
chart_data=chart_data[["date", "M0", "M1", "M2"]]
chart_data=chart_data.sort_values(by="date", ascending=False)
chart_data["M2_pct"]=chart_data["M2"].pct_change(-11)*100
chart_data["M1_pct"]=chart_data["M1"].pct_change(-11)*100
chart_data["M0_pct"]=chart_data["M0"].pct_change(-11)*100
chart_data=chart_data[:24]
px.line(chart_data, x='date', y=["M2_pct", "M1_pct", "M0_pct"])
```

```python
conn=sqlite3.connect('PBOC.db')
df=pd.read_sql_query('SELECT * FROM balance_sheet', conn)
conn.close()
df
```

```python
chart_data=df.copy()
chart_data = chart_data.melt(id_vars='date', var_name='item', value_name='amount')
chart_data['item']=chart_data['item'].str.replace(r'^.*\_', '', regex=True)
selected=["date", "国外资产", "货币黄金", "外汇"]
selected=["date", "国外资产", "外汇"]
chart_data=chart_data[chart_data['item'].isin(selected)]
chart_data['amount']=pd.to_numeric(chart_data['amount']).fillna(0)
chart_data=chart_data[chart_data['amount']!=0]
chart_data=chart_data.groupby(['date', 'item'])['amount'].sum().unstack().reset_index()
chart_data=chart_data.sort_values(by="date", ascending=False)
for col in chart_data.columns:
    if col !='date':
        chart_data[f"{col}_pct"]=chart_data[col].astype(float).pct_change(-11)*100

chart_data=chart_data[:24]
plot_col=[col for col in chart_data.columns if 'pct' in col]
px.line(chart_data, x='date', y=plot_col)
```

```python
conn=sqlite3.connect('china_mof.db')
df=pd.read_sql_query('SELECT * FROM financial_data', conn)
conn.close()
df.columns
```

```python
chart_data=df.copy()
chart_data['date']=pd.to_datetime(chart_data['date'])


def ployt_area(plot_y):
    chart_data=chart_data[plot_y]
    chart_data=chart_data.set_index('date')
    chart_data=chart_data.stack().rename("amount").reset_index().rename(columns={'level_1': 'item'})
    total_per_period = chart_data.groupby('date')['amount'].transform('sum')
    # 2. 測算出百分比
    chart_data['percentage'] = (chart_data['amount'] / total_per_period) * 100
    fig=px.area(chart_data, x='date', y='percentage', color='item',
            #    barmode='relative', 
            )

    fig.update_layout(hovermode="x unified",
                                xaxis=dict(
                                # %Y-%m 代表顯示 2024-12
                                hoverformat="%b-%Y" 
                            )
            )
    fig.show()
```

```python
plot_y=['date', '地方一般公共预算本级收入', '中央一般公共预算收入']
ployt_area(plot_y)
```
